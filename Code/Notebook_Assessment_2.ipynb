````python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv("/content/drive/MyDrive/canada_cpi.csv")

# Columnas relevantes
columnas_cpi = ['All-items', 'Food', 'Shelter', 'Household operations, furnishings and equipment',
                'Clothing and footwear', 'Transportation', 'Gasoline', 'Health and personal care',
                'Recreation, education and reading', 'Alcoholic beverages and tobacco products',
                'All-items excluding food and energy', 'All-items excluding energy',
                'Fresh fruit and vegetables', 'Energy', 'Goods', 'Services']

# Crear  nuevo DataFrame para el crecimiento acumulado
df_cagr = df[columnas_cpi].pct_change(periods=5)

# Calcular el crecimiento acumulado en los últimos cinco años
df_cagr = (df_cagr + 1).pow(1/5) - 1

# Identificar el índice con el mayor crecimiento acumulado en los últimos cinco años
indice_mayor_crecimiento = df_cagr.idxmax(axis=1)

# Añadir columna con el índice de mayor crecimiento al DataFrame original
df['Indice_Mayor_Crecimiento'] = indice_mayor_crecimiento

# Mostrar el resultado
print(df[['Year', 'Indice_Mayor_Crecimiento']])
````

## 3.) Ejercicio (2 punto)

1. Utilizando los valores de los Indices a los precios al consumo de la **Gasolina**, se quiere representar **la evolución anual de sus valores** a lo largo del tiempo junto a sus **medias móviles**. En este caso se mostrarán las medias móviles con ventanas de 2, 4, 6, 8, 10. Es decir cada valor a mostrar en las series móviles (de la media móvil) serán los valores medios en la ventana.

El cálculo de la media móvil debera ser **una función de python**, la cuál se empleara en la creación de las 5 series de ventanas móviles. Los argumentos necesarios serán un dataframe y un entero que represente la ventana a ejecutar. La función devolverá el mismo dataframe pero con una nueva columna que se llame:


```
"gasoline_mean_wnd_" + número de la ventana.
```

Este nombre deberá aparecer además en la legenda.

El gráfico por lo tanto mostrará 5 series de valores ventanas moviles y los valores del CPI de la Gasolina. Pensar cómo **resaltar con el tamaño y el estilo de la linea**, los valores de la Gasolina con respecto al resto...

Trabajar con el gráfico para que sea totalmente entendible, ejes, título, legenda, colores, estilos de las lineas...

````python
#Calculo de la media movil de las 5 series de ventanas.    dataframe con una nueva columna que se llame:   "gasoline_mean_wnd_" + número de la ventana.

# Columnas
Gasolina = df['Gasoline']
Year = df['Year']

# Convertir estas columnas en DataFrames
df1 = pd.DataFrame({'Year': Year})
df2 = pd.DataFrame({'Gasolina': Gasolina})

# Concatenar horizontalmente (axis=1)
Evolucion_anual_gasolina = pd.concat([df1, df2], axis=1)

# Crear una nueva columna 'gasoline_mean_wnd' con la media móvil
ventanas = [2, 4, 6, 8, 10]
Evolucion_anual_gasolina_media_movil= pd.DataFrame() #Creo un Dataframe
for ventana in ventanas:    #Crea una nueva columna ventanas          # Calcula la media movil para la columna"Gasolina"
    Evolucion_anual_gasolina_media_movil[f'gasoline_mean_wnd_{ventana}'] = Gasolina.rolling(window=ventana).mean()
   # print(Evolucion_anual_gasolina_media_movil)
# Concatenar horizontalmente (axis=1) y se crea un nuevo dataframe
Total_Evolucion_anual_gasolina_media_movil= pd.concat([df1, df2,Evolucion_anual_gasolina_media_movil], axis=1)
Total_Evolucion_anual_gasolina_media_movil
````

````python
#  El gráfico por lo tanto mostrará 5 series de valores ventanas moviles y los valores del CPI de la Gasolina. Pensar cómo resaltar con el tamaño y el estilo de la linea, los valores de la Gasolina con respecto al resto...

# Crear la figura y ejes
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

# Graficar las medias móviles
Total_Evolucion_anual_gasolina_media_movil.plot(x='Year', y=['gasoline_mean_wnd_2', 'gasoline_mean_wnd_4', 'gasoline_mean_wnd_6', 'gasoline_mean_wnd_8', 'gasoline_mean_wnd_10'], marker='o', linestyle='dashed', markersize=1, linewidth=2, ax=ax, label=['Wnd_2', 'Wnd_4', 'Wnd_6', 'Wnd_8', 'Wnd_10'])

# Graficar la serie 'Gasolina'
Total_Evolucion_anual_gasolina_media_movil.plot(x='Year', y='Gasolina', marker='o', linestyle='-', markersize=4, linewidth=2, ax=ax, label='Gasolina')

# Configurar etiquetas y título
ax.set_xlabel('Year')
ax.set_ylabel('Valor de Gasolina')
ax.set_title('Evolucion anual y Medias Moviles de Gasolina')

# Mostrar la leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
````
