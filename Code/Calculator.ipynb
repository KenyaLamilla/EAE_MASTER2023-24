```python
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b != 0:
        return a / b
    else:
        return "No se puede dividir por cero."

while True:
    print("\nCalculadora:")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Salir")

    opcion = input("Seleccione qué operación desea hacer (1-5) o escriba el nombre de la operación: ")

    if opcion.lower() in ["1", "sumar"]:
        while True:
            try:
                a = float(input("Ingresa el primer número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el primer numero nuevamente.")

        while True:
            try:
                b = float(input("Ingresa el segundo número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el segundo numero nuevamente.")

        resultado = sumar(a, b)
        print(f"El resultado de sumar {a} y {b} es: {resultado}")

    elif opcion.lower() in ["2", "restar"]:
        while True:
            try:
                a = float(input("Ingresa el primer número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el primer numero nuevamente.")

        while True:
            try:
                b = float(input("Ingresa el segundo número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el segundo numero nuevamente.")

        resultado = restar(a, b)
        print(f"El resultado de restar {b} de {a} es: {resultado}")

    elif opcion.lower() in ["3", "multiplicar"]:
        while True:
            try:
                a = float(input("Ingresa el primer número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el primer numero nuevamente.")

        while True:
            try:
                b = float(input("Ingresa el segundo número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el segundo numero nuevamente.")

        resultado = multiplicar(a, b)
        print(f"El resultado de multiplicar {a} por {b} es: {resultado}")

    elif opcion.lower() in ["4", "dividir"]:
        while True:
            try:
                a = float(input("Ingresa el primer número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el primer numero nuevamente.")

        while True:
            try:
                b = float(input("Ingresa el segundo número: "))
                break
            except ValueError:
                print("Error en valor introducido. No es numérico. Introduce el segundo numero nuevamente.")

        resultado = dividir(a, b)
        if isinstance(resultado, str):
            print(resultado)
        else:
            print(f"El resultado de dividir {a} por {b} es: {resultado}")

    elif opcion.lower() in ["5", "salir"]:
        print("Saliendo de la calculadora.")
        break

    else:
        print("Opción no válida. Por favor, elige una operación válida.")
```
