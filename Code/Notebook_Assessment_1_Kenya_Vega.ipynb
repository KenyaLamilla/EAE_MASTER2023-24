## 3.) Ejercicio (2 puntos)

Utilizando el mismo dataframe, crear una nueva columna llamada: 'delta_pop', que sea el resultado de calcular la diferencia de los valores normalizados con respecto al año siguiente. Para ello, **implementa una función** que calcule el valor siguiente (fila siguiente) y lo reste al actual. Es decir, el valor en el 2015 deberá ser la diferencia entre el siguiente y el actual. Con esta función calculamos el crecimiento (normalizado) de la población año a año.

Acuérdate de documentar la función correctamente.

````python
#creación de nueva columna delta_pop usando una función:

def crecimiento_pob(df_data,columna):
#En la funcion crecmiento_pob() se calcula la diferencia entre el valor de la siguiente fila (siguiente año) menos el actual.
#Se consideran dos objetos dentro de la función:
# 1. df_data: es el dataframe en donde esta la información
# 2. columna: es el nombre de la columna del cual queremos realizar el cálculo
# Para obtener el resultado primero se definieron las dos variables:
  valor_sig=df_data[columna].shift(-1)
# valor_sig: es el valor de la fila(año) siguiente, para obtener este valor usamos la funcion shift(-1) para desplazar hacia arriba una fila
  valor_act=df_data[columna]
# valor_act: es el valor original de la columna
#Con estas variables definidas en el return ponemos la resta de las variables (valor siguiente menos el actual) para obtener el resultado esperado.
  return valor_sig - valor_act
df_data['delta_pop']= crecimiento_pob(df_data,'norm_pop') #creamos la nueva columna con la funcion creada
df_data[['Years','norm_pop','delta_pop']].tail()
````

````python
#Primero determinamos las dos columnas que utilizaremos del df de definiciones que es el código y el nombre:
df_code=df_def[['Code','Indicator Name']]

#Definimos la función imprimir cuyos argumentos seran el df de datos asi como el nuevo df de codigo que creamos con las dos columnas
def imprimir(df_data,df_code):
  #Colocamos un for para que se revisen todas las columnas del dataframe
  for columna in df_data.columns:
    #Creamos la variable media que sera el promedio de cada columna
    media=df_data[columna].mean()
    #Ya que el df de definiciones no contiene los codigos de Years, norm_pop y delta_pop le indicamos que no haga nada si los identifica en el df de datos
    if columna == 'Years' or columna == 'norm_pop' or columna == 'delta_pop' :
      pass
    #Únicamente hay una columna Desempleo que no esta identificada en el df de definiciones, por lo que le indicamos el cálculo a realizar
    elif columna == 'Desempleo':
      #Imprimirá
      resul=print(f'Personas desempleadas \nLa media del campo {columna} es {media:.2f}')
      return resul
    else:
      #Para las columnas que si están dentro de code la función devolverá el nombre del indicador más la media
      defin=df_code[df_code['Code'] == columna]['Indicator Name']
      definicion=defin.iloc[0]
      resultado= print(f'{definicion}\nLa media del campo {columna} es {media:.2f}\n')
  return resultado

#Llamamos a la función indicando los argumentos
imprimir(df_data,df_code)
````

## Ejercicio Extra (1 punto)

En este ejercicio se requiere “predecir” el valor que tomará cada año la población total de la India. Para eso se utilizará el campo normalizado de la población calculado en el ejercicio 2.
La idea es calcular una nueva columna (y_pred_pop) que sea el valor de cada año anterior más la media total (delta_pop calculada en punto 5.)

Es decir, cada año, por ejemplo 2010, puedo predecir el año siguiente (2011) usando el valor de la población (normalizada) en 2010 y sumándole la media aritmética. La media aritmética es el mejor estimador lineal que podemos construir.

Los pasos pueden ser:
1. Hacer una función que sume 2 argumentos, la media y un valor (el valor previo)
2. Usar una función lambda para pasarle a la función, el valor anterior y la medida y generar una nueva columna. Obviamente el primer valor (1991) no se puede calcular y la función devolverá NaN ya que no existe el valor anterior. Pensad en la función shift() para obtener el valor previo de un dataframe.

Para saber cómo de bien hemos predicho cada valor, podemos usar una métrica sencilla que se llama el “mean absolute error” (MAE), es decir el valor medio del error (en valores absolutos). La función para calcular el MAE es:


```
def mae(y, y_pred):
  """
  Función que calcula el mean absolute error o MAE, es decir el valor medio del error (en valores absolutos)
  Argumentos: y: son los valores reales y_pred son los valores predichos
  Return: la función devuelve un float con el valor del calcilo del MAE


  """
  return sum(abs(y-y_pred))
```

Pudiendo obtener el valor de MAE, invocando a la función con **y** (valores reales) y **y_pred** los valores que acabamos de predecir.

¿Cuántos habitantes se ha predicho en el año 2016? Acuérdate en desnormalizar el valor predicho, ya que la predición se realizo con el valor normalizado.


````python
# 1. Crear una función que sume la media de delta_pop y el valor previo
def pred(df_data):
  #Calculamos media de delta_pop
  med_delta_pop=df_data['delta_pop'].mean()
  #Sacamos el valor anterior usando shift
  valor_ant=df_data['norm_pop'].shift(1)
  #el resultado es la suma del valor anterior más la media
  return valor_ant + med_delta_pop
pred(df_data)
#creamos nueva columna con la función
df_data['y_pred_pop']=pred(df_data)
````

````python
#Para calcular el valor del 2016 usamos una máscara
mask=df_data['Years']==2016
df1=df_data[mask]

#Predicción por los datos desnormalizados
pred_2016 = df1['y_pred_pop'] * df_data['POP_TOTL'].max()
pred_2016=pred_2016.values[0]
print(f"El valor predicho para la población en el 2016 es {pred_2016:.2f}")     #:.2f  significa que se imprimirán dos decimales después del punto decimal.
````

````python
"""
Mean Absolute Error, or MAE, is a popular metric.
MAE score is calculated as the average of the absolute error values.
Absolute or abs() is a mathematical function that simply makes a number positive.
Therefore, the difference between an expected and predicted value may be positive or negative and is forced to be positive when calculating the MAE.
MAE = 1 / N * sum for i to N abs(y_i – yhat_i)
"""
#mae de cantidad de personas
def mae(y, y_pred):     # calcula el Error Absoluto Medio
  return sum(abs(y-y_pred))/len(y)

#sin contar el indice 0
y=df_data['POP_TOTL'].iloc[1:]
y_pred=(df_data.iloc[1:]['y_pred_pop'])*df_data['POP_TOTL'].max()
mae(y,y_pred)
````
